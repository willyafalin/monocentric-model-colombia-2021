Construction d'une variable de territoire

********* check if the monocentricity holds *********

********* if not significant : means non monocentric, which makes sense

********* you still capture the size effect

ajouter la photo avec les variables qu'on utilise

```python
total_dataset['DPTO'] = pd.to_numeric(total_dataset['DPTO']) #format numerique


# creer un dictionnaire pour mapper les conditions aux valeurs de 'territory'
# voir notre categorisation 0_main
territory_mapping = {
    ('area', 11): 1,  # bogota: centre
    ('area', (76, 5)): 2,  # cali, medellin: centre
    ('area', (8, 13, 54, 68)): 4,  # grandes aires urbaines et grandes capitales: centre
    ('area', (8, 66, 73, 52, 17, 50, 23)): 6,  # grandes aires urbaines et petites capitales: centre

    # rien ici: bogota cabecera est manquant dans la base 
    ('cabecera', (76, 5)): 3,  # cali, medellin: periurbain
    ('cabecera', (8, 13, 54, 68)): 5,  # grandes aires urbaines et grandes capitales: periurbain
    ('cabecera', (8, 66, 73, 52, 17, 50, 23)): 7,  # grandes aires urbaines et petites capitales: periurbain
    ('cabecera', (15, 41, 20, 47, 19, 44)): 8,  # grandes aires urbaines et sans capitales: periurbain
    ('cabecera', (18, 25, 27, 63, 70)): 10,  # petites aires urbaines et sans capitales: periurbain

    ('rural', (76, 5)): 3,  # cali et medellin: periurbain
    ('rural', (8, 13, 54, 68)): 5,  # grandes aires urbaines et grandes capitales: periurbain
    ('rural', (8, 66, 73, 52, 17, 50, 23)): 7,  # grandes aires urbaines et petites capitales: periurbain
    ('rural', (15, 41, 20, 47, 19, 44)): 9,  # grandes aires urbaines et sans capitales: rural
    ('rural', (18, 25, 27, 63, 70)): 11,  # grandes aires urbaines et sans capitales: rural
}

# appliquer le mapping pour creer la variable 'territory'
def map_territory(row):
    for conditions, value in territory_mapping.items():
        if row['couche_territoriale'] == conditions[0] and (
                (isinstance(conditions[1], int) and row['DPTO'] == conditions[1]) or
                (isinstance(conditions[1], tuple) and row['DPTO'] in conditions[1])):
            return value
    return pd.NA

# lance ta fonction
total_dataset['territory'] = total_dataset.apply(map_territory, axis=1)
```

```python
Loyers reels 

correspond a toute forme de paiement, depenses mensuelles dans le poste de consommation du logement. On prend donc les loyers effectifs, mais aussi les credits des primo accedants. 
On comprend ainsi directement les limites de cette approche, etant donne la faible proportion de primo-accedants dans la base : on ne peut pas prendre en compte le statut de proprieté, on formule des le depart un biais de selection vu qu'on selectionne (quasiment que) les locataires, et non pas les proprietaires. D'un autre coté,
on peut discuter aussi de la pertinence des loyers imputés, car les depenses reelles sont celles qui affectent directement notre cout de la vie, et un proprietaire d'un autre cote jouit de son usufruit, de son cout nul de logement.
```

```python
import numpy as np

# remplacer les valeurs manquantes dans rent_reel_new par mortgagepayment là où rentpayment est manquant
total_dataset['rent_reel'] = np.where(total_dataset['Rentpayment'].isnull(), total_dataset['Mortgagepayment'], total_dataset['Rentpayment'])
total_dataset['rent_reel'] = pd.to_numeric(total_dataset['rent_reel'])

# log transformation
total_dataset['log_rent_reel'] = np.log(total_dataset['rent_reel'].replace(0, np.nan)) #replace pour eviter des pbs avec log(0)

# en tenant compte de l'unite de consommation 
total_dataset['rent_reel_uc'] = total_dataset['rent_reel'] /total_dataset['uc']
total_dataset['log_rent_reel_uc'] = np.log(total_dataset['rent_reel_uc'].replace(0, np.nan)) #replace pour eviter des pbs avec log(0)
```
